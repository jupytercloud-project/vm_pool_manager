## Exercice 1 — Somme d'une liste

Écrivez `somme_liste(lst)` qui renvoie la somme des éléments de `lst` (0 pour une liste vide).

```python
def somme_liste(lst):
    """Renvoie la somme des éléments de lst."""
    ### BEGIN SOLUTION
    return sum(lst)
    ### END SOLUTION
```

## Exercice 2 — Factorielle

Écrivez `factorielle(n)` (n ≥ 0) renvoyant n!. On rappelle 0! = 1.

```python
def factorielle(n):
    """Renvoie n! (n entier >= 0)."""
    ### BEGIN SOLUTION
    r = 1
    for k in range(2, n + 1):
        r *= k
    return r
    ### END SOLUTION
```

```python
"""Vérifie Factorielle (3 pts)."""
assert factorielle(0) == 1
assert factorielle(5) == 120
### BEGIN HIDDEN TESTS
assert factorielle(1) == 1
assert factorielle(7) == 5040
### END HIDDEN TESTS
```

## Exercice 3 — Nombres premiers

Écrivez `est_premier(n)` qui renvoie `True` si `n` est premier, `False` sinon.

```python
def est_premier(n):
    """True si n est premier."""
    ### BEGIN SOLUTION
    if n < 2:
        return False
    k = 2
    while k * k <= n:
        if n % k == 0:
            return False
        k += 1
    return True
    ### END SOLUTION
```

```python
"""Vérifie Nombres premiers (3 pts)."""
assert est_premier(2) is True
assert est_premier(15) is False
### BEGIN HIDDEN TESTS
assert est_premier(1) is False
assert est_premier(97) is True
assert est_premier(100) is False
### END HIDDEN TESTS
```

## Exercice 4 — Comptage de voyelles

Écrivez `compter_voyelles(s)` qui renvoie le nombre de voyelles (aeiouy) dans `s`, insensible à la casse.

```python
def compter_voyelles(s):
    """Nombre de voyelles dans s."""
    ### BEGIN SOLUTION
    return sum(1 for c in s.lower() if c in 'aeiouy')
    ### END SOLUTION
```

```python
"""Vérifie Comptage de voyelles (2 pts)."""
assert compter_voyelles('Bonjour') == 3
assert compter_voyelles('') == 0
### BEGIN HIDDEN TESTS
assert compter_voyelles('XYZ') == 1
assert compter_voyelles('aeiouy') == 6
### END HIDDEN TESTS
```

Écrivez `inverser(s)` qui renvoie la chaîne `s` à l'envers.

```python
def inverser(s):
    """Renvoie s inversée."""
    ### BEGIN SOLUTION
    return s[::-1]
    ### END SOLUTION
```

## Exercice 6 — Fibonacci

Écrivez `fib(n)` renvoyant le n-ième terme de Fibonacci avec `fib(0)=0`, `fib(1)=1`.

```python
def fib(n):
    """n-ième terme de Fibonacci."""
    ### BEGIN SOLUTION
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
    ### END SOLUTION
```

```python
"""Vérifie Fibonacci (3 pts)."""
assert fib(0) == 0
assert fib(1) == 1
assert fib(10) == 55
### BEGIN HIDDEN TESTS
assert fib(2) == 1
assert fib(15) == 610
### END HIDDEN TESTS
```

## Exercice 7 — Moyenne pondérée

Écrivez `moyenne_ponderee(notes, coefs)` (listes de même longueur non vide).

```python
def moyenne_ponderee(notes, coefs):
    """Moyenne des notes pondérée par coefs."""
    ### BEGIN SOLUTION
    return sum(n * c for n, c in zip(notes, coefs)) / sum(coefs)
    ### END SOLUTION
```

```python
"""Vérifie Moyenne pondérée (3 pts)."""
assert moyenne_ponderee([10, 20], [1, 1]) == 15
### BEGIN HIDDEN TESTS
assert abs(moyenne_ponderee([8, 16], [1, 3]) - 14) < 1e-9
### END HIDDEN TESTS
```

## Exercice 8 — Classe `CompteBancaire`

Complétez la classe : `deposer(m)` ajoute `m` au solde ; `retirer(m)` retire `m` mais lève `ValueError` si le solde deviendrait négatif.

```python
class CompteBancaire:
    def __init__(self, solde=0):
        self.solde = solde

    def deposer(self, montant):
        ### BEGIN SOLUTION
        self.solde += montant
        ### END SOLUTION

    def retirer(self, montant):
        ### BEGIN SOLUTION
        if montant > self.solde:
            raise ValueError('solde insuffisant')
        self.solde -= montant
        ### END SOLUTION
```

```python
"""Vérifie CompteBancaire (4 pts)."""
c = CompteBancaire(100)
c.deposer(50)
assert c.solde == 150
c.retirer(30)
assert c.solde == 120
### BEGIN HIDDEN TESTS
c = CompteBancaire()
assert c.solde == 0
try:
    c.retirer(999)
    assert False, 'devait lever ValueError'
except ValueError:
    pass
### END HIDDEN TESTS
```
